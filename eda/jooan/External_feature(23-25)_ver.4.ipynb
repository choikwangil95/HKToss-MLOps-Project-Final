```python
import ast

def extract_last_company_name(labels):
    # NaN 또는 빈 값 처리
    if pd.isna(labels) or labels == '[]' or labels == '':
        return None
    # 문자열인 경우 리스트로 변환
    if isinstance(labels, str):
        try:
            labels_list = ast.literal_eval(labels)
        except Exception:
            return None
    elif isinstance(labels, list):
        labels_list = labels
    else:
        return None
    # 빈 리스트 처리
    if not labels_list:
        return None
    # 마지막 값 반환
    return labels_list[-1]

# 적용
df['stock_name'] = df['stock_list'].apply(extract_last_company_name)

df.info()
```

```python
import pandas as pd
from pykrx import stock

# 2023년 1월 ~ 2025년 12월까지 모든 거래일 리스트 생성
trading_days = []
for year in range(2023, 2026):
    for month in range(1, 13):
        try:
            days = stock.get_business_days(year, month)
            trading_days.extend(days)
        except Exception as e:
            # 월별 데이터가 없을 경우(미래 등) 무시
            continue

# 중복 제거 및 정렬, pandas datetime 변환
trading_days = pd.to_datetime(sorted(set(trading_days)))

def get_next_trading_day(date):
    after = trading_days[trading_days >= date]
    if len(after) > 0:
        return after[0]
    else:
        return pd.NaT
```

```python
offsets = {
    'D_minus_14_date': -14, 'D_minus_7_date': -7, 'D_minus_3_date': -3,
    'D_minus_2_date': -2, 'D_minus_1_date': -1, 'D_day_date': 0,
    'D_plus_1_date': 1, 'D_plus_2_date': 2, 'D_plus_3_date': 3,
    'D_plus_7_date': 7, 'D_plus_14_date': 14
}
date_cols = list(offsets.keys())

# 기준일 생성
for col, offset in offsets.items():
    df[col] = df['wdate'] + pd.to_timedelta(offset, unit='D')
    df[col] = df[col].apply(get_next_trading_day)
```

```python
# 6. 뉴스에 등장하는 코스피 티커만 추출
all_tickers = df['ticker'].dropna().unique().tolist()

# 7. 기준일에 등장하는 모든 날짜 추출
all_dates = pd.concat([df[col] for col in date_cols]).dropna().unique()
all_dates = sorted([d.strftime("%Y%m%d") for d in pd.to_datetime(all_dates)])
```

```python
# 6. pykrx에서 시세/수급 데이터 일괄 수집 및 병합
result_list = []
for ticker in all_tickers:
    # 시세 데이터
    ohlcv = stock.get_market_ohlcv_by_date(all_dates[0], all_dates[-1], ticker)
    # 투자자별 순매수 데이터
    trd = stock.get_market_trading_value_by_date(all_dates[0], all_dates[-1], ticker, detail=True)
    # 컬럼명 자동 확인
    trd_cols = trd.columns.tolist()
    foreign_col = next((c for c in ['외국인', '외국인투자자', '외국인합계'] if c in trd_cols), None)
    individual_col = next((c for c in ['개인', '개인투자자'] if c in trd_cols), None)
    institution_col = next((c for c in ['기관계', '기관', '기관합계', '기관투자자'] if c in trd_cols), None)
    cols_to_use = [c for c in [foreign_col, institution_col, individual_col] if c]
    rename_map = {}
    if foreign_col: rename_map[foreign_col] = 'foreign'
    if institution_col: rename_map[institution_col] = 'institution'
    if individual_col: rename_map[individual_col] = 'individual'
    trd = trd[cols_to_use].rename(columns=rename_map)
    # 병합
    merged = ohlcv.join(trd, how='left')
    merged['ticker'] = ticker
    merged['date'] = merged.index
    result_list.append(merged.reset_index(drop=True))

market_df = pd.concat(result_list, ignore_index=True)
market_df['date'] = pd.to_datetime(market_df['date'])
```
