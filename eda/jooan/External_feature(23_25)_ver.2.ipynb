```python
import ast

def extract_last_company_name(labels):
    # NaN 또는 빈 값 처리
    if pd.isna(labels) or labels == '[]' or labels == '':
        return None
    # 문자열인 경우 리스트로 변환
    if isinstance(labels, str):
        try:
            labels_list = ast.literal_eval(labels)
        except Exception:
            return None
    elif isinstance(labels, list):
        labels_list = labels
    else:
        return None
    # 빈 리스트 처리
    if not labels_list:
        return None
    # 마지막 값 반환
    return labels_list[-1]

# 적용
df['stock_name'] = df['stock_list'].apply(extract_last_company_name)
```

```python
import pandas as pd
from datetime import timedelta
from pykrx import stock
import numpy as np

# 1. 날짜 컬럼 생성
df['wdate'] = pd.to_datetime(df['wdate'])
df['D_minus_30_date'] = df['wdate'] - timedelta(days=30)
df['D_day_date'] = df['wdate']
df['D_plus_1_date'] = df['wdate'] + timedelta(days=1)
df['D_plus_2_date'] = df['wdate'] + timedelta(days=2)
df['D_plus_3_date'] = df['wdate'] + timedelta(days=3)
df['D_plus_7_date'] = df['wdate'] + timedelta(days=7)
df['D_plus_14_date'] = df['wdate'] + timedelta(days=14)
df['D_plus_21_date'] = df['wdate'] + timedelta(days=21)
df['D_plus_30_date'] = df['wdate'] + timedelta(days=30)

# 2. 주말이면 다음 월요일로 보정
def next_monday_if_weekend(date):
    if pd.isna(date):
        return date
    weekday = date.weekday()
    if weekday == 5:  # 토요일
        return date + timedelta(days=2)
    elif weekday == 6:  # 일요일
        return date + timedelta(days=1)
    else:
        return date

cols_to_adjust = ['D_minus_30_date', 'D_plus_1_date', 'D_plus_2_date', 'D_plus_3_date', 'D_plus_30_date']
for col in cols_to_adjust:
    df[col] = df[col].apply(next_monday_if_weekend)

# 3. 필요한 날짜/종목 조합 추출
date_cols = [
    'D_minus_30_date', 'D_day_date', 'D_plus_1_date', 'D_plus_2_date',
    'D_plus_3_date', 'D_plus_7_date', 'D_plus_14_date', 'D_plus_21_date', 'D_plus_30_date'
]
all_dates = set()
for col in date_cols:
    all_dates.update(df[col].dt.strftime('%Y%m%d').tolist())
all_dates = sorted(list(all_dates))
all_tickers = df['ticker'].dropna().unique().tolist()

# 4. OHLCV 데이터 일괄 수집
ohlcv_dict = {}
for ticker in all_tickers:
    try:
        ohlcv = stock.get_market_ohlcv_by_date(
            fromdate=min(all_dates),
            todate=max(all_dates),
            ticker=ticker
        )
        ohlcv_dict[ticker] = ohlcv
    except:
        continue

# 5. 값 추출 함수
def get_ohlcv_val(ticker, date, col):
    if pd.isna(ticker) or pd.isna(date):
        return np.nan
    date_str = date.strftime('%Y%m%d')
    if ticker not in ohlcv_dict:
        return np.nan
    df_ticker = ohlcv_dict[ticker]
    if date_str not in df_ticker.index:
        return np.nan
    return df_ticker.loc[date_str, col]

# 6. 종가 컬럼 생성
df['D-30'] = df.apply(lambda row: get_ohlcv_val(row['ticker'], row['D_minus_30_date'], '종가'), axis=1)
df['close_D-day'] = df.apply(lambda row: get_ohlcv_val(row['ticker'], row['D_day_date'], '종가'), axis=1)
df['D+1'] = df.apply(lambda row: get_ohlcv_val(row['ticker'], row['D_plus_1_date'], '종가'), axis=1)
df['D+2'] = df.apply(lambda row: get_ohlcv_val(row['ticker'], row['D_plus_2_date'], '종가'), axis=1)
df['D+3'] = df.apply(lambda row: get_ohlcv_val(row['ticker'], row['D_plus_3_date'], '종가'), axis=1)
df['D+7'] = df.apply(lambda row: get_ohlcv_val(row['ticker'], row['D_plus_7_date'], '종가'), axis=1)
df['D+14'] = df.apply(lambda row: get_ohlcv_val(row['ticker'], row['D_plus_14_date'], '종가'), axis=1)
df['D+21'] = df.apply(lambda row: get_ohlcv_val(row['ticker'], row['D_plus_21_date'], '종가'), axis=1)
df['D+30'] = df.apply(lambda row: get_ohlcv_val(row['ticker'], row['D_plus_30_date'], '종가'), axis=1)
```

```python
from pykrx import stock

# 필요한 날짜 범위
date_min = df['D_day_date'].min().strftime('%Y%m%d')
date_max = df['D_day_date'].max().strftime('%Y%m%d')
all_tickers = df['ticker'].dropna().unique().tolist()

# 종목별 데이터 캐싱
trading_dict = {}
for ticker in all_tickers:
    try:
        tv_df = stock.get_market_trading_value_by_date(date_min, date_max, ticker)
        trading_dict[ticker] = tv_df
    except:
        continue
```

```python
def fast_trading_value(ticker, date, investor):
    if pd.isna(ticker) or pd.isna(date):
        return np.nan
    date_str = date.strftime('%Y%m%d')
    if ticker not in trading_dict:
        return np.nan
    tv_df = trading_dict[ticker]
    if date_str not in tv_df.index:
        return np.nan
    col_map = {
        '외국인': '외국인합계',
        '기관': '기관합계',
        '개인': '개인'
    }
    return tv_df.loc[date_str, col_map[investor]]

df['foreign_trading'] = df.apply(lambda row: fast_trading_value(row['ticker'], row['D_day_date'], '외국인'), axis=1)
df['institution_trading'] = df.apply(lambda row: fast_trading_value(row['ticker'], row['D_day_date'], '기관'), axis=1)
df['individual_trading'] = df.apply(lambda row: fast_trading_value(row['ticker'], row['D_day_date'], '개인'), axis=1)
```

```python
import yfinance as yf

# USD/KRW 환율 데이터 다운로드
won_dollar = yf.download("KRW=X")

# 최신 환율(종가) 출력
won_dollar.tail()
```

```python
import pandas as pd

# 1. 중복 컬럼 제거
if 'level_0' in won_dollar.columns or ('level_0', '') in won_dollar.columns:
    won_dollar = won_dollar.drop(columns=['level_0'], errors='ignore')

# won_dollar가 MultiIndex 컬럼이면 1단계로 변환
if isinstance(won_dollar.columns, pd.MultiIndex):
    # 예: ('Close', 'KRW=X') → 'Close'
    won_dollar.columns = [col[0] for col in won_dollar.columns]

# Date 인덱스를 컬럼으로 변환
won_dollar = won_dollar.reset_index()  # Date 컬럼 생성

# 날짜 타입 맞추기
df_merged['wdate'] = pd.to_datetime(df_merged['wdate']).dt.date
won_dollar['Date'] = pd.to_datetime(won_dollar['Date']).dt.date

# merge
merged_df = pd.merge(df_merged, won_dollar[['Date', 'Close']], left_on='wdate', right_on='Date', how='left')

# 환율 컬럼명 변경
merged_df.rename(columns={'Close': 'won_dollar'}, inplace=True)
merged_df.drop(columns=['Date'], inplace=True)

# 결과 확인
print(merged_df.head())
```

```python
# 거래량, 등락률 (뉴스 발행일 기준)
merged_df['D-day_volume'] = merged_df.apply(lambda row: get_ohlcv_val(row['ticker'], row['D_day_date'], '거래량'), axis=1)
merged_df['D-day_change'] = merged_df.apply(lambda row: get_ohlcv_val(row['ticker'], row['D_day_date'], '등락률'), axis=1)
```
