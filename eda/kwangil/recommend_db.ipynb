```python
import requests

url = f"http://43.200.17.139:8080/api/v1/userinfo/all"

try:
    response = requests.get(url)
    response.raise_for_status()
    user_datas = response.json()["data"]

    print(user_datas)
except Exception as e:
    print(f"❌ 사용자 정보 조회 실패: {str(e)}")
```
