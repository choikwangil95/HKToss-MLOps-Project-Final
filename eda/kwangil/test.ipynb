```python
df_stocks = pd.read_csv("../../db/news_with_embedding_stocks_formatted.csv")

# stocks decode 해주기
import json


# 문자열 디코딩 함수
def decode_stocks(s):
    try:
        return json.loads(s.encode("utf-8").decode("unicode_escape"))
    except Exception:
        return []


# df 전체에 적용
df_stocks["stocks"] = df_stocks["stocks"].apply(decode_stocks)

df_stocks.head()
```

```python
df = pd.read_csv("../../db/news_3y_with_embeddings.csv")

# 20250101_0001 이렇게 date 칼럼으로 id 생성해주기
df["date2"] = pd.to_datetime(df["date"]).dt.strftime("%Y%m%d")
df["daily_idx"] = df.groupby("date2").cumcount() + 1
df["news_id"] = df["date2"] + "_" + df["daily_idx"].astype(str).str.zfill(4)

df.drop(columns=["date2", "daily_idx"], inplace=True)

# news_id 칼럼을 맨 앞으로 이동
cols = ["news_id"] + [col for col in df.columns if col != "news_id"]
df = df[cols]

df['stocks'] = df_stocks['stocks']  # stocks 컬럼 추가
df.head()
```

```python
# CSV 저장
output_path = '../../db/news_3y.csv' 

df.to_csv(output_path, index=False, encoding='utf-8')
```
