```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

# 임베딩 펼치기 (예: 768차원)
text_emb = df["embedding"].apply(pd.Series).values

# PCA 적합 (전체 차원 유지)
pca = PCA().fit(text_emb)

# 누적 설명 분산
cumsum = np.cumsum(pca.explained_variance_ratio_)

# 예: 누적 95% 설명하는 최소 차원 수
n_components_95 = np.argmax(cumsum >= 0.95) + 1
pca_95 = PCA(n_components=n_components_95)
text_emb_pca_95 = pca_95.fit_transform(text_emb)

print(f"95% 설명하는 차원 수: {n_components_95}")

plt.plot(np.arange(1, len(cumsum) + 1), cumsum)
plt.axhline(0.95, color="r", linestyle="--")
plt.xlabel("차원 수")
plt.ylabel("누적 설명 분산 비율")
plt.title("PCA 설명 분산 누적합")
plt.grid()
plt.show()
```

```python
import torch
import torch.nn as nn

class EmbeddingAutoencoder(nn.Module):
    def __init__(self, input_dim=768, hidden_dim=128):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 512), nn.ReLU(), nn.Linear(512, hidden_dim)
        )
        self.decoder = nn.Sequential(
            nn.Linear(hidden_dim, 512), nn.ReLU(), nn.Linear(512, input_dim)
        )

    def forward(self, x):
        z = self.encoder(x)
        x_hat = self.decoder(z)
        return x_hat, z


# 학습 예시 (단기)
model = EmbeddingAutoencoder()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.MSELoss()

text_emb = df['embedding']

X = torch.tensor(text_emb, dtype=torch.float32)

for epoch in range(1, 101):
    model.train()
    x_hat, z = model(X)
    loss = criterion(x_hat, X)
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()

    if epoch % 10 == 0 or epoch == 1:
        print(f"Epoch {epoch}/100, Loss: {loss.item():.6f}")
```

```python
# 임베딩 압축 결과
with torch.no_grad():
    _, text_emb_ae = model(X)
    text_emb_ae = text_emb_ae.numpy()
```

```python
# 이미 학습한 model 재사용
sim_text_emb = torch.tensor(df["similar_embedding"], dtype=torch.float32)

# encoder만 통과시켜 임베딩 압축
with torch.no_grad():
    sim_text_emb_ae = model.encoder(sim_text_emb).numpy()

# 유사도 계산
from sklearn.metrics.pairwise import cosine_similarity

similarity_ae = cosine_similarity(text_emb_ae, sim_text_emb_ae).diagonal()

# 결과 저장
df["similarity_ae"] = similarity_ae
```
